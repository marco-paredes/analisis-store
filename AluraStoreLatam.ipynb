### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
venta = 'Precio'
ingreso_total_t1 = tienda1[venta].sum()
print(f'Venta total de la tienda1 es: ', ingreso_total_t1)

venta = 'Precio'
ingreso_total_t2 = tienda2[venta].sum()
print(f'Venta total de la tienda2 es: ', ingreso_total_t2)

venta = 'Precio'
ingreso_total_t3 = tienda3[venta].sum()
print(f'Venta total de la tienda3 es: ', ingreso_total_t3)

venta = 'Precio'
ingreso_total_t4 = tienda4[venta].sum()
print(f'Venta total de la tienda4 es: ', ingreso_total_t4)

```

```python
mayor = ingreso_total_t1
if ingreso_total_t2 > ingreso_total_t1:
    mayor = ingreso_total_t2
    print(f"La tienda de mayor venta es la tienda2: {mayor}")
elif ingreso_total_t3 > ingreso_total_t1:
    mayor = ingreso_total_t3
    print(f"La tienda de mayor venta es la tienda3: {mayor}")
elif ingreso_total_t4 > ingreso_total_t1:
    mayor = ingreso_total_t4
    print(f"La tienda de mayor venta es la tienda4: {mayor}")
else:
    print(f"La tienda de mayor venta es la tienda1: {mayor}")
```

# 2. Ventas por categoría

```python
# 2. Agrupamos por tienda y categoría, y contamos cuántas filas hay en cada grupo
producto_categoria_t1 = (
    tienda1.groupby(['Categoría del Producto'])
    .size()              # devuelve una Series con el recuento de filas
)
#para ordenar descendente
producto_categoria_t1 = tienda1['Categoría del Producto'].value_counts()
print(f'TIENDA 1: \n{producto_categoria_t1}')
```

```python
producto_categoria_t2 = (
    tienda2.groupby(['Categoría del Producto'])
    .size()
)
producto_categoria_t2 = tienda2['Categoría del Producto'].value_counts()
print(f'TIENDA 2: \n{producto_categoria_t2}')
```

```python
producto_categoria_t3 = (
    tienda3.groupby(['Categoría del Producto'])
    .size()
)
producto_categoria_t3 = tienda3['Categoría del Producto'].value_counts()
print(f'TIENDA 3: \n{producto_categoria_t3}')
```

```python
producto_categoria_t4 = (
    tienda4.groupby(['Categoría del Producto'])
    .size()
)
producto_categoria_t4 = tienda4['Categoría del Producto'].value_counts()
print(f'TIENDA 4: \n{producto_categoria_t4}')
```

# 5. Envío promedio por tienda

```python
promedio_costo_envio_t1 = tienda1['Costo de envío'].mean()
print(f'Promedio de costo de envío TIENDA 1: {round(promedio_costo_envio_t1,2)}')

promedio_costo_envio_t2 = tienda2['Costo de envío'].mean()
print(f'Promedio de costo de envío TIENDA 2: {round(promedio_costo_envio_t2,2)}')

promedio_costo_envio_t3 = tienda3['Costo de envío'].mean()
print(f'Promedio de costo de envío TIENDA 3: {round(promedio_costo_envio_t3,2)}')

promedio_costo_envio_t4 = tienda4['Costo de envío'].mean()
print(f'Promedio de costo de envío TIENDA 4: {round(promedio_costo_envio_t4,2)}')
```

```python
# Calculamos promedios
#promedio_costo_envio_t1 = tienda1['Costo de envío'].mean()
#promedio_costo_envio_t2 = tienda2['Costo de envío'].mean()

# Preparamos las porciones y etiquetas
sizes  = [promedio_costo_envio_t1, promedio_costo_envio_t2, promedio_costo_envio_t3, promedio_costo_envio_t4]
labels = [f'Tienda 1 ({promedio_costo_envio_t1:.2f})', f'Tienda 2 ({promedio_costo_envio_t2:.2f})', f'Tienda 3 ({promedio_costo_envio_t3:.2f})', f'Tienda 4 ({promedio_costo_envio_t4:.2f})']

# Colores y explode opcional para resaltar la primera porción
colors  = ['#66c2a5','#fc8d62', '#8da0cb', '#e78ac3']
explode = (0.1, 0,0,0)

# Creamos la torta
plt.figure(figsize=(6,6))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',   # muestra % de cada porción
    startangle=90,       # rotación inicial
    colors=colors,
    explode=explode,
    shadow=True
)
plt.title('Comparación de costo de envío promedio')
plt.axis('equal')        # asegura círculo perfecto
plt.show()
```
